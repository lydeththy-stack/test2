```python
import csv

SIZE = 50
hash_table = [None] * SIZE

def hash_code(key):
    return key % SIZE

def insert(key, value):
    index = hash_code(key)
    while hash_table[index] is not None:
        index = (index + 1) % SIZE
    hash_table[index] = (key, value)

def search(key):
    index = hash_code(key)
    while hash_table[index] is not None:
        if hash_table[index][0] == key:
            return hash_table[index]
        index = (index + 1) % SIZE
    return None

# Load CSV
with open("/content/drive/MyDrive/students.csv") as f:
    reader = csv.DictReader(f)
    for row in reader:
        insert(int(row["StudentID"]), row["Name"])

# Test search
print(search(1001))
```
